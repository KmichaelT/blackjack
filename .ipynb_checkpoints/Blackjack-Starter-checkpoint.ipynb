### Rules:

`1. ` The game will have two players: the Dealer and the Player. The game will start off with a deck of 52 cards. The 52 cards will consist of 4 different suits: Clubs, Diamonds, Hearts and Spades. For each suit, there will be cards numbered 1 through 13. <br>
**Note: No wildcards will be used in the program**

`2. ` When the game begins, the dealer will shuffle the deck of cards, making them randomized. After the dealer shuffles, it will deal the player 2 cards and will deal itself 2 cards from. The Player should be able to see both of their own cards, but should only be able to see one of the Dealer's cards.
 
`3. ` The objective of the game is for the Player to count their cards after they're dealt. If they're not satisfied with the number, they have the ability to 'Hit'. A hit allows the dealer to deal the Player one additional card. The Player can hit as many times as they'd like as long as they don't 'Bust'. A bust is when the Player is dealt cards that total more than 21.

`4. ` If the dealer deals the Player cards equal to 21 on the **first** deal, the Player wins. This is referred to as Blackjack. Blackjack is **NOT** the same as getting cards that equal up to 21 after the first deal. Blackjack can only be attained on the first deal.

`5. ` The Player will never see the Dealer's hand until the Player chooses to 'stand'. A Stand is when the player tells the dealer to not deal it anymore cards. Once the player chooses to Stand, the Player and the Dealer will compare their hands. Whoever has the higher number wins. Keep in mind that the Dealer can also bust. 

```python
import random
class BlackJack():
  
    def __init__(self,player):
        suits=['spades','hearts','clubs','diamonds']
        values=[i for i in range(1,14)]
        self.deck={(f"{v} of {s}"):v for v in values for s in suits }
        self.name_of_card=list(self.deck.keys())
        self.player=player
        self.player_card = random.sample(self.name_of_card, 2)
        self.dealer_card = random.sample(self.name_of_card, 2)
        self.dealer_total=self.deck[self.dealer_card[0]] + self.deck[self.dealer_card[1]]
        self.player_total=self.deck[self.player_card[0]] + self.deck[self.player_card[1]]
        
    def hit(self):
        while self.player_total<22:
            newcard=random.sample(self.name_of_card, 1)
            self.player_card.append(newcard)
            self.player_total+=self.deck[newcard[0]]
            print (f"you're given a {newcard[0]} new total is {self.player_total}")
            if self.player_total>21:
                print(f"BUST!! you lost! you have ${int(self.bet)-int(self.bet)}")

            break
    def stand(self):
        print(f"my cards are {self.dealer_card[0]}, {self.dealer_card[1]} which is {self.dealer_total}")
        while self.dealer_total<17:
            newcard=random.sample(self.name_of_card, 1)
            self.dealer_card.append(newcard)
            self.dealer_total+=self.deck[newcard[0]]
            print(f"i got a {newcard[0]} new total is {self.dealer_total}")
            if self.dealer_total>21:
                print(f"BUST!! you won! you have ${int(self.bet)+int(self.bet)}")
                self.player_total=0
            break
    def compare(self):
        while self.dealer_total<= 21 and self.player_total<= 21:
            if self.dealer_total > self.player_total:
                print(f"you lost! you have ${int(self.bet)-int(self.bet)}")
                
            
            elif self.dealer_total < self.player_total:
                print(f"you won! you have ${int(self.bet)+int(self.bet)}")
                
            break
    def play(self):
        
        self.bet=input(f'hello {self.player},\nhow much do you want to bet for?  ')
        print(f"my cards are {self.dealer_card[0]}, Hidden")
        print(f"your cars are {self.player_card[0]} and {self.player_card[1]} which is {self.player_total}")
        if self.player_total==21:
            print("BLACK JACK")
        while self.dealer_total<= 21 and self.player_total<= 21:
            choice=input('do you want to hit or stand?')
            if choice=="stand":
                print('you chose to stand \n')
                self.stand()
                print('_________________\n')
                break
            elif choice=="hit" :
                print('you chose to hit\n')
                self.hit()
                print('_________________\n')
            elif self.bet==0:
                break
        self.compare()
        
def main():
    player=input("who is playing ")
    game=BlackJack(f'{player}')
    game.play()
    while True:
        startover=(input("do you want to start over? y/n ")).lower()
        if startover=='y':
            game=BlackJack(f'{player}')
            game.play()
        elif startover=='n':
            break
        else:
            print("please enter y or n")
                
main()
```
